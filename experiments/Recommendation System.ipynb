```python
tfidf_vector=TfidfVectorizer()
feature_vector=tfidf_vector.fit_transform(df['genres'])
print(feature_vector)
```

```python
cos_similarity=cosine_similarity(feature_vector)
print(cos_similarity)
```

```python
print(cos_similarity.shape)
```

```python
title_list=df['title'].tolist()
print(title_list)
```

## Getting the movie name by user:-

```python
movie_name=input("Enter the movie name :")
```

```python
close_match=difflib.get_close_matches(movie_name,title_list)
print(close_match)
```

```python
index_movie=df[df['title']==close_match[0]]['index'].values[0]
print(index_movie)
```

```python
similarity_score=list(enumerate(cos_similarity[index_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key= lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
print("Movies suggested for you : \n")
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if (i<=10):
        print(i,'.',title_from_index)
    i+=1        
```
